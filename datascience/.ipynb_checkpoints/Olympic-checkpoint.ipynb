```python

import pandas as pd
import numpy as np
olympic=pd.read_excel('olympic.xlsx',sheetname='ALL MEDALISTS',skiprows=4)
olympicDf=pd.DataFrame()
country=np.unique(olympic['NOC'])


for i in country:
    maleGold=len(olympic[(olympic['Gender']=='Men') & (olympic['Medal']=='Gold') & (olympic['NOC']==i)])
    maleSilver=len(olympic[(olympic['Gender']=='Men') & (olympic['Medal']=='Silver') & (olympic['NOC']==i)])
    maleBronze=len(olympic[(olympic['Gender']=='Men') & (olympic['Medal']=='Bronze') & (olympic['NOC']==i)])
    femaleGold=len(olympic[(olympic['Gender']=='Women') & (olympic['Medal']=='Gold') & (olympic['NOC']==i)])
    femaleSilver=len(olympic[(olympic['Gender']=='Women') & (olympic['Medal']=='Silver') & (olympic['NOC']==i)])
    femaleBronze=len(olympic[(olympic['Gender']=='Women') & (olympic['Medal']=='Bronze') & (olympic['NOC']==i)])
    data=np.array([[i,'Male',maleGold,maleSilver,maleBronze],['','Female', femaleGold,femaleSilver,femaleBronze]])
    
    columns=['Country','Category','Gold','Silver','Bronze']
    dataframe = pd.DataFrame(data=data,columns=columns)
    olympicDf=olympicDf.append(dataframe,ignore_index=True)
```

```python
olympicDf
```

```python
olympicDf.groupby(by='Country').head()
```
