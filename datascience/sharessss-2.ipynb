```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
url = "http://www.merolagani.com/CompanyList.aspx"
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text,'lxml')
```

```python
raw_data = soup.findAll('div',attrs={'id':'collapse_0'})
```

```python
company_symbols = []
company_urls = []
for data in raw_data:
    company_links = data.findAll('a')
    for symbol in company_links:
        company_symbols.append(symbol.string.strip(""))
        url ='http://www.merolagani.com/CompanyDetail.aspx?symbol={}'.format(symbol.string.strip(""))
        company_urls.append(url)
#         print(symbol.string.strip(""))
```

```python
values = []
i = 0 ## repalce with 0 to get all 
while i < len(company_urls):
    req = requests.get(company_urls[i])
    soup = BeautifulSoup(req.text,'lxml')
    tables = soup.findChildren('table',attrs={'class':'table-zeromargin'})
    company_table = tables[0]
    data = company_table.findChildren(['td'])
    for td in data:
        span = td.findChildren(['span'])
        strong = td.findChildren(['strong'])
        for data in strong:
            if data.text:
                values.append(data.text)
            
                print(data.text)
        for data in span:
            values.append(data.text)
            print(data.text)
        if td.string:
            text = td.string.replace("0\xa0-\xa0\r\n","")
            text = text.replace("\r\n",'')
            text = text.strip("\n")
            text = text.strip(" ")
            print(text)
            if text:
                values.append(text)
            
        else:
            values.append("NULL")
            print("Empty")
#     print(soup.text)
    print("\n***********************************************************************\n")
    
    i+=1
```
